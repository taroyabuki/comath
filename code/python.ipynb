# 9 多次元の確率分布

```
X = DiscreteUniform('X', range(1, 7))
Y = X % 4
{(x, y): P(And(Eq(X, x), Eq(Y, y))) for x in range(1, 7) for y in range(4)}
```

```
X = DiscreteUniform('X', range(1, 7))
Y = DiscreteUniform('Y', range(1, 7)) % 4 # Xと独立な確率変数Y
```

```
{(x, y): P(And(Le(X, x), Le(Y, y))) for x in range(1, 7) for y in range(4)}
```

```
(E(X), E(Y),               # 平均
 variance(X), variance(Y), # 分散
 std(X), std(Y),           # 標準偏差
 covariance(X, Y),         # 共分散
 correlation(X, Y))        # 相関係数
```

```
uX, uY = E(X), E(Y); sX, sY = std(X), std(Y)
(E(X), E(Y),                       # 平均
 E((X - uX)**2), E((Y - uY)**2),   # 分散
 E((X - uX) * (Y - uY)),           # 共分散
 E((X - uX) * (Y - uY) / sX / sY)) # 相関係数
```

```
print(sum((x * P(Eq(X, x)) for x in range(1, 7))))         # Xの平均
print(sum((x - uX) * (y - uY) * P(And(Eq(X, x), Eq(Y, y))) # 共分散
          for x in range(1, 7) for y in range(4)))
```

```
X = DiscreteUniform('X', range(1, 7)); Y = X % 4
[P(And(X <= x, Y <= y)) for x in range(1, 7) for y in range(4)] == \
[P(X <= x) * P(Y <= y) for x in range(1, 7) for y in range(4)]
```

```
U = DiscreteUniform('X', range(1, 7)); X = U % 2; Y = U % 3
[P(And(X <= x, Y <= y)) for x in range(2) for y in range(3)] == \
[P(X <= x) * P(Y <= y) for x in range(2) for y in range(3)]
```

```
A = Matrix([[1, 0], [1, 1], [0, 1]])
G = A.T @ A # 1                                              # 1
vals = G.eigenvals(multiple=True)                            # 2-1
vecs = [v for e in G.eigenvects() for v in e[2]]             # 2-2
s = [sqrt(x) for x in vals if x > 0]; S = diag(*s)           # 3
r = len(s)                                                   # 4
V = Matrix.hstack(*GramSchmidt(vecs[-r:], orthonormal=True)) # 5
U = A @ V @ S.inv()                                          # 6
U, S, V.T, U @ S @ V.T
```

```
A = np.array([[1, 0], [1, 1], [0, 1]])
G = A.T @ A                                             # 1
vals, vecs = linalg.eig(G)                              # 2
s = [np.sqrt(x) for x in vals if x > 0]; S = np.diag(s) # 3
r = len(s)                                              # 4
V = vecs[:, :r]                                         # 5
U = A @ V @ linalg.inv(S)                               # 6
U, S, V.T, U @ S @ V.T
```
